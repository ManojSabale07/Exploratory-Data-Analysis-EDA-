# Travel Data Analysis

In this EDA project, we'll embark on a journey through the world of travel data. Behind every memorable trip—from the flight taken to the hotel booked—lies a vast amount of information that can tell us a powerful story about human behavior and preferences.

For travel agencies, tour operators, and even individual travelers, making the right decisions isn't a matter of chance. It requires a deep understanding of trends, popular destinations, and customer booking patterns. This is where data analysis becomes an invaluable tool.

Our goal in this Exploratory Data Analysis (EDA) project is to act as data explorers. We will clean, analyze, and visualize the travel dataset to uncover valuable insights into travel trends. By exploring this data, we hope to gain a better understanding of what drives people's travel choices, helping to make future journeys smoother and more personalized.

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('ProdTaken', axis = 1)
y = df['ProdTaken']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 1)
```

```python
cats_feature = X.select_dtypes(include = 'object').columns
cats_feature
```

```python
nums_feature = X.select_dtypes(exclude = 'object').columns
nums_feature
```

## Scaling --> Numerical Data
## Encoding --> Categorical Data

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

### ColumnTransformer
Applies transformations to the columns of an array/pandas dataframe

```python
numeric_transformer = StandardScaler()
OH_transformer = OneHotEncoder(drop = 'first')
preprocessor = ColumnTransformer(
    [
        ('OneHotEncoder', OH_transformer, cats_feature),
        ('StandardScaler', numeric_transformer, nums_feature)
    ])
```

```python
preprocessor
```

```python
X_train = preprocessor.fit_transform(X_train)
pd.DataFrame(X_train)
```

```python
X_test = preprocessor.transform(X_test)
X_test
```
